```python
import os
import subprocess
import sys

PROJECT_PATH = os.path.expanduser("~/mlops_project")


if not os.path.exists(PROJECT_PATH):
    os.makedirs(PROJECT_PATH)

def initialize_git():
    os.chdir(PROJECT_PATH)
    subprocess.run(["git", "init"])
    print("Initialized Git Repository.")

def create_virtual_env():
    os.chdir(PROJECT_PATH)
    subprocess.run([sys.executable, "-m", "venv", "venv"])
    print(" Virtual Environment Created.")

def install_packages():
    os.chdir(PROJECT_PATH)
    subprocess.run(["venv/bin/pip", "install", "--upgrade", "pip"])
    

    required_packages = ["tensorflow", "numpy", "pandas", "scikit-learn", "matplotlib"]
    subprocess.run(["venv/bin/pip", "install"] + required_packages)
    
    with open("requirements.txt", "w") as f:
        subprocess.run(["venv/bin/pip", "freeze"], stdout=f)

    print(" Dependencies Installed & `requirements.txt` Created.")

GIT_REMOTE_URL = "git@github.com:your_username/mlops-project.git"

def push_to_git():
    os.chdir(PROJECT_PATH)
    subprocess.run(["git", "add", "."])
    subprocess.run(["git", "commit", "-m", "Initial MLOps Deployment"])
    subprocess.run(["git", "branch", "-M", "main"])
    subprocess.run(["git", "remote", "add", "origin", GIT_REMOTE_URL])
    subprocess.run(["git", "push", "-u", "origin", "main"])
    print(" Project Pushed to GitHub.")

def pull_and_setup():
    os.chdir(PROJECT_PATH)
    subprocess.run(["git", "pull", "origin", "main"])
    subprocess.run(["venv/bin/pip", "install", "-r", "requirements.txt"])
    print(" Project Pulled & Dependencies Installed.")

def run_model():
    os.chdir(PROJECT_PATH)
    model_script = os.path.join(PROJECT_PATH, "run_model.py")

    with open(model_script, "w") as f:
        f.write("""
import tensorflow as tf
import numpy as np

# Load the trained model
model = tf.keras.models.load_model('/mnt/data/tic-tac-toe.keras')

# Dummy input data (Randomized)
X_sample = np.random.rand(1, 9)

# Make prediction
prediction = model.predict(X_sample)
print(f"Predicted Output: {prediction}")
""")

    subprocess.run(["venv/bin/python", model_script])
    print("✅ Model Successfully Tested.")

initialize_git()
create_virtual_env()
install_packages()
push_to_git()
pull_and_setup()
run_model()
```
